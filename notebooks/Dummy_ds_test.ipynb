```python
image_df = ds_bag.get_table_as_dataframe('Image')
```

```python
image_df
```

```python
image_rids = image_df.RID.tolist()
```

```python
import numpy as np

np.random.shuffle(image_rids)

train_count = int(len(image_rids)*0.2)
val_count = int(len(image_rids)*0.05)
test_count = int(len(image_rids)*0.03)

training_rids = image_rids[:train_count]
test_rids = image_rids[train_count:train_count + test_count]
validation_rids = image_rids[train_count + test_count:train_count + test_count + val_count]
```

```python
example_ds = execution.create_dataset(['LAC'], description='A Test dataset')
training_dataset = execution.create_dataset(['LAC', 'Training'], description='A Test train dataset')
test_dataset = execution.create_dataset(['LAC', 'Test'], description='A Test test dataset')
validation_dataset = execution.create_dataset(['LAC', 'Validation'], description='A Test validation datset')
```

```python
EA.add_dataset_members(dataset_rid=training_dataset, members=training_rids)
EA.add_dataset_members(dataset_rid=test_dataset, members=test_rids)
EA.add_dataset_members(dataset_rid=validation_dataset, members=validation_rids)
```
